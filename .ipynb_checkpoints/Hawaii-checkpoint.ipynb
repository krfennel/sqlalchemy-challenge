```python
# from sqlalchemy import create_engine
# from sqlalchemy import Column, Integer, String, Float

# from sqlalchemy.ext.declarative import declarative_base
# from sqlalchemy.ext.automap import automap_base

import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine
```

```python
engine = create_engine("postgresql://postgres:postgres@localhost:5432/SQLalchemy-challenge")
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python

data = engine.execute("SELECT * FROM station")
for record in data:
    print(record)
```

```python
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
data = engine.execute("   \
set datestyle to 'SQL, MDY';  \
select to_char(cast(date as date),'MM/DD/YYYY') as date, prcp   \
from measurement   \
where cast(date as date) > cast('08/23/2016' as date)\
order by  \
  cast(date as date)  \
")

```

```python
for record in data:
    print(record)
```
